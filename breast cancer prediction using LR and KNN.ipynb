```python
y=df['diagnosis']
x=df.drop(['diagnosis'],1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_sc=sc.fit_transform(x)
```

```python
x_sc.shape
```

```python
from sklearn.linear_model import LogisticRegression
from warnings import filterwarnings
filterwarnings('ignore')
from sklearn.model_selection import cross_val_score,GridSearchCV
lr=LogisticRegression()
```

```python
score=cross_val_score(lr,x_sc,y,cv=5,scoring='roc_auc')
print("bias error:",(1-np.mean(score)),"variance error:",np.std(score,ddof=1))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
param={'n_neighbors':np.arange(1,100),'weights':['uniform','distance']}
gs=GridSearchCV(knn,param,cv=5,scoring='roc_auc')
gs.fit(x_sc,y)
```

```python
gs.best_params_
```

```python
knn=KNeighborsClassifier(n_neighbors=25,weights='distance')
score=cross_val_score(knn,x_sc,y,cv=5,scoring='roc_auc')
score.mean() ,score.std(ddof=1)
```

```python
print("bias error:",1-score.mean(),"variance error:",score.std(ddof=1))
```
